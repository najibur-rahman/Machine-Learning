```python

from tensorflow import keras
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
# module data vag kora train and test er jonno
df = keras.datasets.fashion_mnist
type(df)
```

```python
(x_train, y_train), (x_test, y_test) = df.load_data()
```

```python
# min-max scailer kora hoise
x_train = x_train/255
y_train = y_train/255
```

```python
x_train[0]
```

```python
# flattened mane 2d picture ke 1d te convert kora
x_train_flattened = x_train.reshape(len(x_train), 28*28)
x_test_flattened = x_test.reshape(len(x_test), 28*28)
```

```python
x_train_flattened.shape
```

```python
# keras sequensila class e thake
from keras.models import Sequential
from keras.layers import Dense, Flatten
```

```python
# data ready kora
model = Sequential()
model.add(Flatten(input_shape=[28,28])) # input layer
model.add(Dense(100, activation="relu")) # hidden layer with 100 neurons
model.add(Dense(10, activation='softmax')) # output layer
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# total 1200 but eikane 960 ta kortese cz of 20% test er jonno raktese
history = model.fit(
    x_train,
    y_train,
    epochs=10,
    batch_size=50,
    verbose=1,
    validation_split=0.2
)
```

```python
y_train.shape
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred = y_pred.argmax(axis= -1)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```
