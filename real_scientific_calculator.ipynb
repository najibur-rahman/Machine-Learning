SCIENTIFIC CALCULATOR

```python
def add(a, b):
  result= a+b
  print(result)

def substract(a, b):
  result= a-b
  print(result)

def multiplication(a, b):
  result= a*b
  print(result)

def division(a, b):
  result= a/b
  print(result)

def modulus(a, b):
  result= a%b
  print(result)

def root(a):
  result= a**0.5
  print(result)

def square(a):
  result= a**2
  print(result)

def cube(a):
  result= a*a*a
  print(result)

def cubic_root(a):
  result= a**(1/3)
  print(result)

def Percentage(num, percent):
  result = (a * percent)/100
  print(percent, "percent in", a, "is", result)


while True:
  print("Enter your input like as operand, operator, operand")
  a = int(input())
  operator = input()

  if operator == "root" or operator == "square" or operator == "cube" or operator == "cubic root":
    if operator == "root":
      root(a)
    if operator == "square":
      square(a)
    if operator == "cube":
      cube(a)
    if operator == "cubic root":
      cubic_root(a)
  else:
    b = int(input())
    if operator == "+":
      add(a, b)
    elif operator == "-":
      substract(a, b)
    elif operator == "*":
      multiplication(a, b)
    elif operator == "/":
      division(a, b)
    elif operator == "modulus":
      modulus(a, b)
    elif operator == "%":
      Percentage(a, b)
    else:
      print("Wrong input. Please try again...")

  print("do you continue: (y/n)")
  test = input()
  if(test == 'n'):
    break
```
